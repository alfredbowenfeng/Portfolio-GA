```python
import tushare as ts
import pandas as pd
import time, datetime
ts.set_token('add25d81ad873db57d60c417e8e5fcad1bd6085b492979ee5821b751')
pro = ts.pro_api()
```

```python
startdate, enddate = '20150101', '20190430'
benchmarkdf = pd.read_csv('Fund-tushare-20151231-20190331.csv')
symboldf = benchmarkdf.drop_duplicates(subset=['symbol'], keep='first')
symboldf = symboldf.sort_values('symbol', ascending=True)
symbollist = symboldf['symbol'].tolist()
```

```python
def getdf(cc):
    df = pro.query('daily', ts_code=cc, start_date=startdate, end_date=enddate)
    df = df.drop('pre_close', axis=1, inplace=False)
    df = df.drop('change', axis=1, inplace=False)
    df = df.drop('pct_chg', axis=1, inplace=False)
    df = df.drop('vol', axis=1, inplace=False)
    df = df.drop('amount', axis=1, inplace=False)
    df = df.sort_values('trade_date', ascending=True)
    return df

def mergedf(olddf, newdf):
    df = pd.concat([olddf, newdf], axis=0, ignore_index=True)
    return df
```

```python
sdf = pd.DataFrame(columns=['ts_code', 'trade_date', 'open', 'high', 'low', 'close'])

for s in symbollist:
    try:
        df = getdf(s)
    except:
        print('sleep for 65 seconds with ' + s)
        time.sleep(65)
        df = getdf(s)
    sdf = mergedf(sdf, df)
    print(s + '    ', end='\r')
```
