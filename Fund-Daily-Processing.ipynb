```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.style.use(style='ggplot')
plt.rcParams['figure.figsize'] = (20, 12)
```

```python
dailydf = pd.read_csv('Daily-tushare-20151231-20190430.csv')
dailydf = dailydf.drop('Unnamed: 0', axis=1, inplace=False)
dailydf0101 = pd.read_csv('Daily-tushare-20150101-20190430.csv')
dailydf0101 = dailydf0101.drop('Unnamed: 0', axis=1, inplace=False)
funddf = pd.read_csv('Fund-tushare-20151231-20190331.csv')
funddf = funddf.drop('Unnamed: 0', axis=1, inplace=False)
daterange = funddf.sort_values('end_date', ascending=True)
daterange = daterange.drop_duplicates(subset=['end_date'], keep='first')['end_date']
fddf = pd.DataFrame(dailydf['trade_date'].values.tolist(), columns=['trade_date'])
fddf = fddf.drop_duplicates(subset=['trade_date'], keep='first')
fplist = []
```

```python
def decideRange(date):
    dt = daterange.values.tolist()
    dr = []
    for x in dt:
        if x % 10000 == 1231:
            dr.append(x)
        elif x % 10000 == 630:
            dr.append(x)
        else:
            pass
    while dr[-1] > date:
        dr.pop()
    return dr[-1]

def decideDate(code, date):
    dd = dailydf0101[dailydf0101['ts_code']==code]
    if len(dd[dd['trade_date']==date]) == 0:
        while date >= 20150101:
            if len(dd[dd['trade_date']==date]) == 0:
                if date % 10000 > 1231:
                    date = (date // 10000)*10000 + 1231
                else:
                    date += -1
            else:
                dD = date
                break
    else:
        dD = date
    return dD
```

```python
for date in fddf['trade_date']:
    weights = funddf[funddf['end_date']==decideRange(date)]
    pricelist, weightlist = [], []
    for s in weights['symbol']:
        price = dailydf[dailydf['ts_code']==s]
        try:
            pricelist.append(price[price['trade_date']==date]['close'].values[0])
        except:
            try:
                d = decideDate(s, date)
                price = dailydf0101[dailydf0101['ts_code']==s]
                pricelist.append(price[price['trade_date']==d]['close'].values[0])
            except:
                price = weights[weights['symbol']==s]['mkv'].values[0]/weights[weights['symbol']==s]['amount'].values[0]
                pricelist.append(price)
    total = np.multiply(np.array(weights['amount'].values.tolist()), pricelist).sum()
    value = np.multiply(np.array(weights['amount'].values.tolist()), pricelist).tolist()
    for v in value:
        weightlist.append(v/total)
    fp = np.multiply(np.array(weightlist), np.array(pricelist)).sum()
    fplist.append(fp)
    print(str(date), end='\r')
```

```python
changelist = []
for i in range(0, len(fplist)):
    if i != 0:
        changelist.append((fplist[i]-fplist[i-1])/fplist[i-1])
    else:
        changelist.append(0)
```

```python
fddf.insert(1, 'value', fplist)
fddf.insert(2, 'change', changelist)
```

```python
fddf.to_csv('Fund-Daily-Semi-20151231-20190430.csv')
```

```python
fvlist = []
for date in fddf['trade_date']:
    weights = funddf[funddf['end_date']==decideRange(date)]
    pricelist, volatilitylist, weightlist = [], [], []
    for s in weights['symbol']:
        price = dailydf[dailydf['ts_code']==s]
        try:
            pricelist.append(price[price['trade_date']==date]['close'].values[0])
            volatility = price[price['trade_date']==date]
            volatilitylist.append((volatility['high'].values[0] - volatility['low'].values[0])/(volatility['close'].values[0]))
        except:
            try:
                d = decideDate(s, date)
                price = dailydf0101[dailydf0101['ts_code']==s]
                pricelist.append(price[price['trade_date']==d]['close'].values[0])
                volatility = price[price['trade_date']==date]
                volatilitylist.append((volatility['high'].values[0] - volatility['low'].values[0])/(volatility['close'].values[0]))
            except:
                price = weights[weights['symbol']==s]['mkv'].values[0]/weights[weights['symbol']==s]['amount'].values[0]
                pricelist.append(price)
                volatilitylist.append(0)
    total = np.multiply(np.array(weights['amount'].values.tolist()), pricelist).sum()
    value = np.multiply(np.array(weights['amount'].values.tolist()), pricelist).tolist()
    for v in value:
        weightlist.append(v/total)
    fvw = np.array(weightlist).tolist()
    fvv = np.array(volatilitylist).tolist()
    fv = 1
    for i in range(0, len(fvw)):
        fv = fv * fvw[i] * fvv[i]
        fv = fv ** (1/len(fvw))
    fvlist.append(fv)
    print(str(date), end='\r')
```
