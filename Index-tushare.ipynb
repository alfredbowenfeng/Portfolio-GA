```python
import tushare as ts
import pandas as pd
import time, datetime
ts.set_token('add25d81ad873db57d60c417e8e5fcad1bd6085b492979ee5821b751')
pro = ts.pro_api()
```

```python
inputSDate = ['20190101', '20180701', '20180101', \
              '20170701', '20170101', '20160701', '20160101']
inputEDate = ['20190430', '20181231', '20180630', \
              '20171231', '20170630', '20161231', '20160630']
industrylis = []
indexdf = pd.DataFrame(columns=['index_code', 'con_code', \
                                'trade_date', 'weight'])

def getdf(inputSDate, inputEDate):
    df = pro.index_weight(index_code='399300.SZ', \
                          start_date=inputSDate, end_date=inputEDate)
    return df

def mergedf(olddf, newdf):
    df = pd.concat([olddf, newdf], axis=0, ignore_index=True)
    return df
```

```python
for i in range(0, len(inputSDate)):
    idf = getdf(inputSDate[i], inputEDate[i])
    indexdf = mergedf(indexdf, idf)
```

```python
industry = pro.stock_basic(exchange='', list_status='L', \
                           fields='ts_code,industry')

for i in range(0, len(indexdf)):
    try:
        industrylis.append(industry[industry['ts_code']==indexdf['con_code'][i]]['industry'].values[0])
        print('success ' + str(i+1) + '/' + str(len(indexdf)) + '    ', end='\r')
    except:
        industrylis.append('N/A')
        print('failure ' + str(i+1) + '/' + str(len(indexdf)) + '    ', end='\r')
```

```python
indexdf.insert(3, 'industry', industrylis)
indexdf = indexdf.sort_values('trade_date', ascending=True)
indexdf
```
