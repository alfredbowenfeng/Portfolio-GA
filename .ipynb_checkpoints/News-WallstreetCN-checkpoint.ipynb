```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
import json, time, datetime, pkuseg
import pandas as pd
from requests_html import HTMLSession

session = HTMLSession()
seg = pkuseg.pkuseg(postag=True)
```

```python
baseurl1 = 'https://api.wallstreetcn.com/apiv1/search/live?cursor='
baseurl2 = '&channel=global-channel&limit=20&start_time=1451577600&end_time=1556640000&score=0'
L = ['display_time', 'content_text']
contentlis = []
```

```python
def getPage(cursor):
    page = json.loads(session.get(baseurl1 + str(cursor) + baseurl2).text)
    return(page)

def getNews(tn):
    t = tn['display_time']
    n = tn['content_text'][:-1]
    l = [t, n]
    return l
```

```python
for i in range(1, 11137):
    try:
        page = getPage(i)
    except:
        print('sleep for 30s when i=' + str(i))
        time.sleep(30)
        page = getPage(i)
    for j in range(0, len(page['data']['items'])):
        tn = page['data']['items'][j]
        l = getNews(tn)
        L.append(l)
    print(str(i) + '/11136' + '    ', end='\r')
```

```python
dfL = pd.DataFrame(data=L[2:], columns=[L[0], L[1]])
dfL
```

```python
for i in range(0, len(dfL)):
    try:
        contentlis.append(seg.cut(dfL['content_text'][i]))
    except:
        contentlis.append([])

dfL_seg = dfL.drop('content_text', axis=1, inplace=False)
dfL_seg.insert(1, 'content_text_seg', contentlis)
dfL_seg = dfL_seg.sort_values('display_time', ascending=True)
```

```python
dfL_seg.to_csv('News-WallstreetCN-Seg-20160101-20190430.csv')
```
