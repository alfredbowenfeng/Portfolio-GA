```python
import tushare as ts
import pandas as pd
import time
ts.set_token('add25d81ad873db57d60c417e8e5fcad1bd6085b492979ee5821b751')
pro = ts.pro_api()
```

```python
datelist = ['20151231', '20160630', '20161231', '20170630', '20171231', '20180630', '20181231']
benchmarkdf = pd.read_csv('Index-tushare-20160101-20190415.csv')
con_codedf = benchmarkdf.drop_duplicates(subset=['con_code'], keep='first')
con_codedf = con_codedf.sort_values('con_code', ascending=True)
con_codelist = con_codedf['con_code'].tolist()
mbdf = pd.DataFrame(columns=['ts_code', 'end_date', 'bz_item', 'bz_sales', 'bz_profit', 'bz_cost', 'curr_type'])
```

```python
def mergedf(olddf, newdf):
    df = pd.concat([olddf, newdf], axis=0, ignore_index=True)
    return df
```

```python
for con_code in con_codelist:
    for date in datelist:
        try:
            df = pro.fina_mainbz(ts_code=con_code, period=date, type='P')
        except:
            print('sleep 65 seconds with ' + con_code)
            time.sleep(65)
        mbdf = mergedf(mbdf, df)
    print(con_code + '    ', end='\r')
```
